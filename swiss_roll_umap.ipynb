```python
import numpy as np
from sklearn.datasets import make_swiss_roll
import umap.umap_ as umap
import matplotlib.pyplot as plt

# Generate the 3D Swiss Roll dataset
n_samples = 10000
noise = 0.05
X, color = make_swiss_roll(n_samples=n_samples, noise=noise)

# Plot the 3D Swiss Roll in a separate figure
fig1 = plt.figure(figsize=(8, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax1.set_title("3D Swiss Roll")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")
plt.show()

# Perform UMAP dimensionality reduction to 2D
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
X_embedded = reducer.fit_transform(X)

# Plot the 2D embedding in a separate figure
fig2 = plt.figure(figsize=(8, 6))
ax2 = fig2.add_subplot(111)
scatter = ax2.scatter(X_embedded[:, 0], X_embedded[:, 1], c=color, cmap=plt.cm.Spectral)
ax2.set_title("2D UMAP Embedding")
ax2.set_xlabel("UMAP 1")
ax2.set_ylabel("UMAP 2")
plt.colorbar(scatter, ax=ax2, label="Color")
plt.show()

# also umap 1 component
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=1, random_state=42)
X_embedded = reducer.fit_transform(X)
```
